# Gráficos

```julia
using Plots
x = 1:10
y = rand(10, 2)
plot(x,y,title="two lines",label=["line 1" "line 2"],lw=3)
```

```julia
y = rand(10, 4)
plot(x,y, layout=(4,1))
```

```julia
plot(x,y, layout=(2,2))
```

```julia
p1 = plot(x,y) # 4 linhas
p2 = scatter(x,y) # Só pontos
p3 = plot(x,y, xlabel="este tem label", lw=3, title="subtitulo")
p4 = histogram(x, y[:,1])

plot(p1, p2, p3, p4, layout=(2,2), legend=false)
```
